```python
#loadfile
fs=open('fly.txt','r')

mydictionary={}

#create loop to strip and split string into a list 
for lines in fs:
    fields=lines.strip('\n').split()
    
    #make conditional to print all with DROME 
    if 'DROME'in lines:
        flybase=fields[-1]
        ac=fields[-2]
        print(flybase + '\t' + ac)
        
        #add output to dictionary
        mydictionary[flybase]=ac
        
fs.close()
```
